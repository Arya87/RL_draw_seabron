```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
plt.style.use('ggplot')
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
# 读取实验日志
df_sac = pd.read_csv('./sac.csv')
df_sac_cheetah = df_sac.query('env == "cheetah_run"')
df_sac_walker = df_sac.query('env == "walker_run"')
df_sac_quadruped = df_sac.query('env == "quadruped_run"')
df_sac_fish = df_sac.query('env == "fish_swim"')
df_sac_pendulum = df_sac.query('env == "pendulum_swingup"')
df_sac_hopper = df_sac.query('env == "hopper_hop"')


df_cheetah = pd.read_csv('./human/cheetah-run/eval.csv')
df_walker = pd.read_csv('./human/walker-run/eval.csv')
df_quadruped = pd.read_csv('./human/quadruped-run/eval.csv')
df_fish = pd.read_csv('./human/fish-swim/eval.csv')
df_hopper = pd.read_csv('./human/hopper-hop/eval.csv')
df_pendulum = pd.read_csv('./human/pendulum-swingup/eval.csv')

df_per_cheetah = pd.read_csv('./SAC_PER/cheetah-run/eval.csv')
df_per_walker = pd.read_csv('./SAC_PER/walker-run/eval.csv')
df_per_quadruped = pd.read_csv('./SAC_PER/quadruped-run/eval.csv')
df_per_fish = pd.read_csv('./SAC_PER/fish-swim/eval.csv')
df_per_hopper = pd.read_csv('./SAC_PER/hopper-hop/eval.csv')
df_per_pendulum = pd.read_csv('./SAC_PER/pendulum-swingup/eval.csv')
```

```python
# 大画布尺寸 长：18，高：10
f = plt.figure(figsize=(12, 7))
plt.rcParams.update({'font.size': 11})


# 调色盘
enmax_palette = ["#CC0000", "#FF9900", "#00CD00", "#1e90ff"]
color_codes_wanted = ['red', 'orange', 'green', 'blue']
c = lambda x: enmax_palette[color_codes_wanted.index(x)]
# sns.set_style("ticks",{"axes.facecolor":'#FFFFFF'})

# 子画布（第一个）
# 参数1：子位图总行数 参数2：子位图总列数  参数3：子位图所在位置
'''
相关参数说明
    -sns.set_style("ticks",{"axes.facecolor":'#FFFFFF',"axes.linewidth": 5}): 设置图表样式
    -plt.grid(b=True, ls=':', color='grey', lw=1)：表格内线段类型
    -plt.title('Cheetah Run')：表名
    -plt.xlabel(' ')：x轴名
    -plt.ylabel('Episode return'): y轴名
    -plt.xlim(0,1000000)：x轴数据范围
    -plt.ylim(0,1000)：y轴数据范围
    -ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.1f}'.format(x/1000000)))：设置坐标轴样式

'''
f.add_subplot(231)
sns.set_style("ticks",{"axes.facecolor":'#FFFFFF',"axes.linewidth": 5})
plt.grid(b=True, ls=':', color='grey', lw=1)
plt.title('Cheetah Run')
plt.xlabel(' ')
plt.ylabel('Episode return')
plt.xlim(0,1000000)
plt.ylim(0,1000)
ax = sns.lineplot(x="step", y="episode_reward", color=c("blue"), lw=2, data=df_sac_cheetah)
sns.lineplot(x="step", y="episode_reward", color=c("red"), lw=3, data=df_cheetah)
sns.lineplot(x="step", y="episode_reward", color=c("orange"), lw=2, data=df_per_cheetah)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.1f}'.format(x/1000000)))


# 自定义图例
plt.plot([], [], color=c("red"), label='SAC-HIFER', lw=3)
plt.plot([], [], color=c("orange"), label='SAC-PER')
plt.plot([], [], color=c("blue"), label='SAC')
plt.legend(loc='lower right', frameon=True, fontsize=12)



# 加粗外框线
for spine in ax.spines.values():
    spine.set_linewidth(1.5)


# 子画布（2行3列第2个）
f.add_subplot(232)
sns.set_style("ticks",{"axes.facecolor":'#FFFFFF',"axes.linewidth": 5})
plt.grid(b=True, ls=':', color='grey', lw=1)
plt.title('Walker Run')
plt.xlabel(' ')
plt.ylabel(' ')
plt.xlim(0,1000000)
plt.ylim(0,1000)
ax = sns.lineplot(x="step", y="episode_reward", color=c("blue"), lw=2, data=df_sac_walker)
sns.lineplot(x="step", y="episode_reward", color=c("red"), lw=3, data=df_walker)
sns.lineplot(x="step", y="episode_reward", color=c("orange"), lw=2, data=df_per_walker)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.1f}'.format(x/1000000)))
# 加粗外框线
for spine in ax.spines.values():
    spine.set_linewidth(1.5)


# 子画布（2行3列第3个）
f.add_subplot(233)
sns.set_style("ticks",{"axes.facecolor":'#FFFFFF',"axes.linewidth": 5})
plt.grid(b=True, ls=':', color='grey', lw=1)
plt.title('Quadruped Run')
plt.xlabel(' ')
plt.ylabel(' ')
plt.xlim(0,1000000)
plt.ylim(0,1000)
ax = sns.lineplot(x="step", y="episode_reward", color=c("blue"), lw=2, data=df_sac_quadruped)
sns.lineplot(x="step", y="episode_reward", color=c("red"), lw=3, data=df_quadruped)
sns.lineplot(x="step", y="episode_reward", color=c("orange"), lw=2, data=df_per_quadruped)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.1f}'.format(x/1000000)))
# 加粗外框线
for spine in ax.spines.values():
    spine.set_linewidth(1.5)


# 子画布（2行3列第4个）
f.add_subplot(234)
sns.set_style("ticks",{"axes.facecolor":'#FFFFFF',"axes.linewidth": 5})
plt.grid(b=True, ls=':', color='grey', lw=1)
plt.title('Hopper Hop')
plt.xlabel('Environment steps (×10⁶)')
plt.ylabel('Episode return')
plt.xlim(0,1000000)
plt.ylim(0,500)
ax = sns.lineplot(x="step", y="episode_reward", color=c("blue"), lw=2, data=df_sac_hopper)
sns.lineplot(x="step", y="episode_reward", color=c("red"), lw=3, data=df_hopper)
sns.lineplot(x="step", y="episode_reward", color=c("orange"), lw=2, data=df_per_hopper)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.1f}'.format(x/1000000)))
# 加粗外框线
for spine in ax.spines.values():
    spine.set_linewidth(1.5)


# 子画布（2行3列第5个）
f.add_subplot(235)
sns.set_style("ticks",{"axes.facecolor":'#FFFFFF',"axes.linewidth": 5})
plt.grid(b=True, ls=':', color='grey', lw=1)
plt.title('Pendulum Swingup')
plt.xlabel('Environment steps (×10⁶)')
plt.ylabel(' ')
plt.xlim(0,1000000)
plt.ylim(0,1000)
ax = sns.lineplot(x="step", y="episode_reward", color=c("blue"), lw=2, data=df_sac_pendulum)
sns.lineplot(x="step", y="episode_reward", color=c("red"), lw=3, data=df_pendulum)
sns.lineplot(x="step", y="episode_reward", color=c("orange"), lw=2, data=df_per_pendulum)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.1f}'.format(x/1000000)))
# 加粗外框线
for spine in ax.spines.values():
    spine.set_linewidth(1.5)


# 子画布（2行3列第6个）
f.add_subplot(236)
sns.set_style("ticks",{"axes.facecolor":'#FFFFFF',"axes.linewidth": 5})
plt.grid(b=True, ls=':', color='grey', lw=1)
plt.title('Fish Swim')
plt.xlabel('Environment steps (×10⁶)')
plt.ylabel(' ')
plt.xlim(0,1000000)
plt.ylim(0,1000)
ax = sns.lineplot(x="step", y="episode_reward", color=c("blue"), lw=2, data=df_sac_fish)
sns.lineplot(x="step", y="episode_reward", color=c("red"), lw=3, data=df_fish)
sns.lineplot(x="step", y="episode_reward", color=c("orange"), lw=2, data=df_per_fish)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.1f}'.format(x/1000000)))
# 加粗外框线
for spine in ax.spines.values():
    spine.set_linewidth(1.5)

# 子图间隔
plt.tight_layout()
plt.show()
```
